```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
```

```python
print(x)
print(y)
```

```python
plt.scatter(x,y) # Scatter Plot
```

```python
# create a scatter plot in betwwen x, y using alpha, colorcode, marker grid line
plt.figure(figsize=(6,2))
plt.scatter(x,y, c="#223BDC", alpha=0.9, marker="X")
```

```python
# also represent the xlabel, ylabel and title within the above programm
plt.figure(figsize=(6,2))
plt.scatter(x,y, c="#223BDC", alpha=0.9, marker="X")
plt.xlabel("this is my x axis")
plt.ylabel("this is my y axis")
plt.title("this is my x vs y")
plt.grid()
```

```python
x = np.linspace(1,10 , 100)
y=np.sin(x)
```

```python
print(x)
print(y)
```

```python
plt.scatter(x,y) # sinusoidal wave
```

```python
# Represent the above code in good manner

plt.figure(figsize=(8,2))
plt.plot(x,y , '--r')
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my sin wave graph")
plt.grid()
```

```python
x = ['a', 'b' , 'c','d','e']
y = np.random.rand(5)

print(x)

print(y)
```

```python
plt.figure(figsize=(5,3))
plt.bar(x,y,color = '#7E10D5')   #Bar Chart
plt.xlabel("representing my cat data")
plt.ylabel("representing my num data")
plt.title("this is my bar plot")
plt.grid()
```

```python
plt.figure(figsize=(5,3))
plt.barh(x,y,color = '#7E10D5')  #Bar Chart with horizontal data representation
plt.xlabel("representing my cat data")
plt.ylabel("representing my num data")
plt.title("this is my bar plot")
plt.grid()
```

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
plt.figure(figsize=(6,3))
plt.hist(data,color="#700EBE")   # Histogram plot
plt.xlabel("representing uniqe val")
plt.ylabel("representing feq")
plt.title("Hist plot example")
plt.grid()
plt.savefig("test.png")
```

```python
# Create a 3-d chart using the scatter plot

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection="3d")
ax.scatter(x,y,z,c="r",alpha=0.9)
plt.xlabel("my x axis")
plt.ylabel("my y axis")
plt.show()
```
