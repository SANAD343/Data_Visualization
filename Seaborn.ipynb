# Seaborn : Statistical Data Visualization

* Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

# Source : https://seaborn.pydata.org/

image.png

```python
import seaborn as sns
```

# The Iris Dataset

* This data sets consists of 3 different types of irises (Setosa, Versicolour, and Virginica) petal and sepal length, stored in a 150x4 numpy.ndarray

* The rows being the samples and the columns being: Sepal Length, Sepal Width, Petal Length and Petal Width.

```python
iris=sns.load_dataset("iris")
iris
```

```python
iris.head(5)
```

```python
# scatter plot between sepal_length and sepal_width
sns.scatterplot(x= iris.sepal_length, y=iris.sepal_width)
```

```python
# scatter plot between sepal_lenth vas petal_length
sns.scatterplot(x = iris.sepal_length , y = iris.petal_length)
```

```python
# print a displot for sepal length

sns.displot(iris["sepal_length"])
```

```python
# load a new dataset ("tips" dataset for more plotting)

tips=sns.load_dataset("tips")
```

```python
print(tips)
```

```python
tips
```

```python
# scatter plot between total bill vas total tip

sns.scatterplot(x=tips.total_bill, y=tips.tip)
```

```python
tips.head()
```

```python
# pull data for smokers
tips['smoker'].value_counts()
```

```python
tips['size'].value_counts()
```

```python
# relational plot between total bill and tip with respect to the smoker
sns.relplot(x = tips.total_bill , y = tips.tip , data = tips,hue = 'smoker' )
```

```python
# relational plot between total bill and tip with respect to the size

sns.relplot(x = tips.total_bill , y = tips.tip , data = tips,style=  'size' )
```

```python
sns.relplot(x = tips.total_bill , y = tips.tip , data = tips,style=  'size',hue = 'time' )
```

```python
tips
```

```python
# CatPlot between day and total_bill

sns.catplot(x="day", y="total_bill", data=tips)
```

```python
sns.catplot(x = 'smoker' , y = 'tip' , data = tips )
```

```python
sns.jointplot(x = tips.total_bill , y = tips.tip)
```

```python
# Pair Plot within the tips dataset
sns.pairplot(tips)
```

```python
# Pairplot within the iris dataset
sns.pairplot(iris)
```
